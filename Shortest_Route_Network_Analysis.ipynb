```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
```

```python
ox.__version__
```

```python
place = 'Kathmandu Metropolitan, Kathmandu, Nepal'
```

```python
data = ox.graph_from_place(place, network_type='drive')
```

```python
fig, ax = ox.plot_graph(data, bgcolor = 'white', node_color = 'red', node_size =1, edge_color='grey')
```

```python
# Finding the shortest route
start_point = (27.442, 85.436)
end_point = (27.821, 85.365)
```

```python
start_node = ox.distance.nearest_nodes(data, start_point[1], start_point[0])
end_node = ox.distance.nearest_nodes(data, end_point[1], end_point[0])
```

```python
shortest_path = nx.shortest_path(data, start_node, end_node, weight='length')
fig, ax =ox.plot_graph_route(data, shortest_path, route_linewidth=4, bgcolor='white', node_size=2, edge_color='yellow')
```
